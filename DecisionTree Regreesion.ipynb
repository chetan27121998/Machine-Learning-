### DecisionTree Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Position_Salaries.csv")
df
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.shape
```

```python
# Selecting the Input and Output from the dataset.
X = df[["Level"]]     # Input
Y = df["Salary"]     # Output
```

```python
#Apply DecisionTreeRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
# Create the object of DecisionTreeRegressor
dcr = DecisionTreeRegressor(criterion='mse',random_state=0)
```

```python
# Train the model
dcr.fit(X,Y)
```

```python
# Testing the model
Y_pred = dcr.predict([[6.5]])
```
